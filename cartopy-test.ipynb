```python
import matplotlib.pyplot as plt
from matplotlib import ticker as mticker
import cartopy 
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
from numpy import arange
```

```python
cent_long=25
box=[-180, 180, -90, 90]
axesfont=12
grid_w = -120
grid_e = 180
grid_s = -90
grid_n = 90
lon_step=60
lat_step=30
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude = cent_long))

ax.set_extent(box, crs=ccrs.PlateCarree())

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, alpha=0.65, linestyle='--')

gl.top_labels = False

gl.right_labels = False

gl.xlabel_style = {'size': axesfont}

gl.ylabel_style = {'size': axesfont}

gl.xlocator = mticker.FixedLocator(arange(grid_w, grid_e + 1., lon_step))

gl.ylocator = mticker.FixedLocator(arange(grid_s, grid_n + 1., lat_step))

gl.xformatter = LONGITUDE_FORMATTER

gl.yformatter = LATITUDE_FORMATTER

ax.coastlines()
```
